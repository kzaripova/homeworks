```python
pd.read_sql(
    """
    SELECT s.id as id,
    w.№_скважины as №_скважины,
    g.ГТМ as ГТМ,
    p.тип_насоса as тип_насоса,
    p.производительность as производительность,
    p.напор as напор,
    r.Q_жидкости as Q_жидкости,
    r.Q_нефти as Q_нефти
    FROM skvazhiny AS s
    LEFT JOIN well AS w ON w.id = s.well_id
    LEFT JOIN gte AS g ON g.id = s.gte_id
    LEFT JOIN pumps AS p ON p.id = s.pumps_id
    LEFT JOIN products AS r ON r.id = s.products_id
    """,
    sql)
```

```python
query = """
SELECT w.№_скважины as №_скважины,
    r.Q_жидкости as Q_жидкости,
    r.Q_нефти as Q_нефти,
    round(100 - Q_нефти/Q_жидкости*100) AS обводнённость
    FROM skvazhiny AS s
    LEFT JOIN well AS w ON w.id = s.well_id
    LEFT JOIN products AS r ON r.id = s.products_id
    GROUP BY обводнённость,w.№_скважины

"""

pd.read_sql(query, sql)
```
