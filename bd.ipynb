```python
pd.read_sql(
    """
    SELECT b.Skvazhina as Skvazhina,
    a.Pump as Pump,
    g.GTE as GTE,
    o.OST as Oil_saturation_thickness,
    b.pressure,
    b.water
    FROM skvazhiny AS b
    LEFT JOIN Pump AS a ON a.id = b.pump_id
    LEFT JOIN GTE AS g ON g.id = b.gte_id
    LEFT JOIN oil AS o ON o.id = b.oil_id
    """,
    sql)
```
